```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
def html_request(url):
    
    response = requests.get(url)
    print("Your response status code is: ", response.status_code)
    
    soup = BeautifulSoup(response.content, "html.parser")
    
    return soup
```

#### Retrieve an arbitrary Wikipedia page of "Python" and create a list of links on that page url ='https://en.wikipedia.org/wiki/Python'.

```python
html_request("https://en.wikipedia.org/wiki/Python")
```

```python
all_links = soup.find_all("a")
all_links
```

#### Create a Python list with the top ten FBI's Most Wanted names: url = 'https://www.fbi.gov/wanted/topten'

```python
top_most_wanted = html_request("https://www.fbi.gov/wanted/topten")
```

```python
top_most_wanted
```

```python
top_most_wanted.select(".title")
```

```python
most_wanted_list = []

for name in top_most_wanted.select(".title"):
    most_wanted_list.append(name.get_text().strip())
```

```python
most_wanted_list
```
