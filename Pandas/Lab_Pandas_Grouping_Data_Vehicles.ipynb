Group by the data by the Make  using count function

```python
cars.groupby("Make").count()
```

What brand has the most cars?

```python
cars["Make"].value_counts().idxmax()
```

<b>show the average CO2_Emission_Grams/Km  by Brand

```python
# I will convert Grams/Mile to Grams/Km

cars["CO2 Emission Grams/Km"] = cars["CO2 Emission Grams/Mile"] / 1.60934

cars
```

```python
# And now calculating the average by brand.

brand_co2 = cars.groupby("Make")["CO2 Emission Grams/Km"].apply(lambda x: x.mean() )
```

```python
brand_co2
```

<b>show the average CO2_Emission_Grams/Km  by Brand ... sorted

```python
brand_co2.sort_values()
```

Use `pd.cut` or `pd.qcut` to create 4 groups (bins) of cars, by Year. We want to explore how cars have evolved decade by decade.

```python
cars['Year'].describe()
```

```python
cars["Decade"] = pd.cut(x = cars["Year"], bins = [1980, 1990, 2000, 2010, 2020], labels = ["80's", "90's", "00's", "10's"]) 

cars.Decade.unique()
```

### Did cars consume more gas in the eighties?

show the average City_Km/Liter by year_range

```python
# I will convert City MPG to Km/Liter

cars["City Km/Liter"] = cars["City MPG"] * 0.42514

cars
```

```python
# Calculating the usage for every decade.

cars.groupby("Decade")["City Km/Liter"].apply(lambda x: x.mean() )
```

Which brands are more environment friendly?

```python
env_friendl = cars.groupby(["Decade", "Make"])["CO2 Emission Grams/Km"].apply(lambda x: x.mean())
env_friendl
```

Does the drivetrain affect fuel consumption?

```python
# Converting Highway MPG to Km/Liter

cars["Highway Km/Liter"] = cars["Highway MPG"] * 0.42514
cars
```

```python
# Calculating the mean for every type of drivetrain.

cars.groupby("Drivetrain")["Highway Km/Liter", "City Km/Liter"].apply(lambda x: x.mean())
```

Do cars with automatic transmission consume more fuel than cars with manual transmission?

```python
# Changing the values for transmission column, so that they will be more clear - automatic or manual.

def transmission_renaming(x):
    if x.startswith("A"):
        return "Automatic"
    else:
        return "Manual"
```

```python
cars["Transmission"] = list(map(transmission_renaming, cars["Transmission"]))
```

```python
cars["Transmission"].unique()
```

```python
# Calculating the fuel consumption mean for both transmission type.

cars.groupby("Transmission")["Highway Km/Liter", "City Km/Liter"].apply(lambda x: x.mean())
```

Use `groupby` and `aggregate` with different aggregation measures for different columns:

aggregate with average City_Km/Liter and the count of the Trans

```python
cars.groupby("Transmission").agg({"City Km/Liter" : ["mean"], "Transmission" : ["count"]})
```

aggregate with average City_Km/Liter and the minimum of the Trans

```python
cars.groupby("Transmission").agg({"City Km/Liter" : ["mean", "min"]})
```
